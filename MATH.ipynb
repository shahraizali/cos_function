## Cosine Function

$\frac{1-cos(2x)}{x^2}$



```python
import matplotlib.pyplot as plt
```

```python

# Import libraries 
import matplotlib.pyplot as plt 
import numpy as np 
import math
  
# Creating vectors X and Y 
x = np.linspace(-0.0000001, 0.0000001, 500) 
y = (1-np.cos(2*x))/(x**2)

# 
fig = plt.figure(figsize = (20, 5))


plt.grid(True)
# Create the plot 
plt.plot(x, y) 
  
# Show the plot 
plt.show() 

```
